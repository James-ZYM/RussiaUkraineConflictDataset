```python
#path = '//Submissions//2022-02-24//CombatFootage.csv'
path = 'C://Users//kathr//OneDrive//Documents//GitHub//NLPexam//Submissions//2022-02-24//CombatFootage.csv'

df = pd.read_csv(path)

submissions = df['title'].rename('submission')
split_file_name = os.path.normpath(path)
split_file_name = split_file_name.split(os.sep)

sub_reddit = pd.Series([split_file_name[9].split('.')[0]] * len(submissions), name = 'sub_reddit') # change index
date = pd.Series([split_file_name[8].split('.')[0]] * len(submissions), name = 'date') # change index

print(date)
```

```python
language = []
for i in submissions:
    try:
        language.append(detect(i))
    except:
        language.append("none")

print(list(set(language)))
language = pd.Series(language, name = 'language')
```

```python
df_dict = {'submission':submissions, 'sub_reddit': sub_reddit, 'date': date, 'language': language}

dataframe = pd.DataFrame(df_dict)
dataframe
```
